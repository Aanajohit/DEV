```python
# STEP 1: Install required libraries
# Run these commands in terminal if not already installed:
# pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# STEP 2: Create a DataFrame from the data
data = {
    'Name': ['John', 'Jane', 'Mike', 'Emily', 'Alex'],
    'Age': [25, 30, 22, 28, 35],
    'Score': [90, 85, 78, 95, 88]
}
df = pd.DataFrame(data)

# STEP 3: Display the DataFrame
print("DataFrame:")
print(df)

# STEP 4: Data Analysis
print("\nData Analysis:")
print("Mean Age:", df['Age'].mean())
print("Maximum Score:", df['Score'].max())
print("Minimum Score:", df['Score'].min())

# STEP 5: Data Visualization
plt.figure(figsize=(8, 4))
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Scores of Students')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load with proper header row
df = pd.read_excel('kpop_idols.csv.xlsx', header=1)

# Display the first few rows to confirm
print(df.head())
```

```python
# Clean column names (lowercase, no spaces)
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

print("Updated columns:")
print(df.columns)
```

```python
import numpy as np
import pandas as pd

# Fix: Make the array rectangular (3x3), with consistent lengths
data = np.array([['', 'Col1', 'Col2'],
                 ['Row1', 1, 2],
                 ['Row2', 3, 4]])

# Slice to build DataFrame: data[1:, 1:] = values, data[1:, 0] = row labels, data[0, 1:] = column labels
df1 = pd.DataFrame(data=data[1:, 1:], index=data[1:, 0], columns=data[0, 1:])
print("DataFrame from structured array:\n", df1, "\n")

# 2D NumPy array to DataFrame
my_2darray = np.array([[1, 2, 3], [4, 5, 6]])
df2 = pd.DataFrame(my_2darray)
print("DataFrame from 2D NumPy array:\n", df2, "\n")

# Dictionary to DataFrame (keys become columns)
my_dict = {1: ['1', '3'], 2: ['1', '2'], 3: ['2', '4']}
df3 = pd.DataFrame(my_dict)
print("DataFrame from dictionary:\n", df3, "\n")

# Create DataFrame from list with a custom column name
my_df = pd.DataFrame(data=[4, 5, 6, 7], index=range(0, 4), columns=['A'])
print("Single-column DataFrame:\n", my_df, "\n")

# Series to DataFrame (default: keys become index)
my_series = pd.Series({"UnitedKingdom": "London",
                       "India": "NewDelhi",
                       "United States": "Washington",
                       "Belgium": "Brussels"})
df4 = pd.DataFrame(my_series, columns=["Capital"])
print("DataFrame from Series:\n", df4, "\n")

# Shape and length of DataFrame
df5 = pd.DataFrame(np.array([[1, 2, 3], [4, 5, 6]]))
print("Shape of DataFrame:", df5.shape)
print("Number of rows (len):", len(df5.index))
```

```python
import pandas as pd
data = {
'sender': ['alice@example.com', 'bob@example.com', 'alice@example.com'],
'receiver': ['bob@example.com', 'alice@example.com', 'carol@example.com'],
'subject': ['Hello', 'Meeting Reminder', 'Project Update'],
'timestamp': ['2023-08-01 10:00:00', '2023-08-02 14:30:00', '2023-08-03 09:15:00'],
'content': ['Hi Bob,\n\nHow are you?', 'Hi Alice,\n\nDon\'t forget the meeting at 3 PM.',
'Hi Carol,\n\nHere\'s the latest project update.'] }
df = pd.DataFrame(data)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.to_csv('emails.csv', index=False)
print("CSV file created successfully.")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("emails.csv")
print(df.info())
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.dropna(inplace=True)
df['email_length'] = df['content'].apply(len)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='email_length', bins=30, kde=True)
plt.xlabel('Email Length')
plt.ylabel('Count')
plt.title('Distribution of Email Lengths')
plt.show()
top_senders = df['sender'].value_counts()[:10]
top_receivers = df['receiver'].value_counts()[:10]
plt.figure(figsize=(12, 6))
sns.barplot(x=top_senders.index, y=top_senders.values)
plt.xticks(rotation=45)
plt.xlabel('Sender')
plt.ylabel('Number of Emails')
plt.title('Top 10 Email Senders')
plt.tight_layout()
plt.show()
df['year_month'] = df['timestamp'].dt.to_period('M')
email_activity = df.groupby('year_month').size()
plt.figure(figsize=(12, 6))
email_activity.plot(kind='line')
plt.xlabel('Year-Month')
plt.ylabel('Number of Emails')
plt.title('Email Activity Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
